```python
from chromadb import PersistentClient

def get_chunks_by_source_and_page(source_path, page_number, collection_name="rag2", db_path="./chroma_db"):
    client = PersistentClient(path=db_path)
    collection = client.get_collection(collection_name)

    # Proper filter using $and operator
    results = collection.get(
        where={
            "$and": [
                {"source": source_path},
                {"page": page_number}
            ]
        },
        include=["documents", "metadatas"]
    )

    documents = results.get("documents", [])
    metadatas = results.get("metadatas", [])

    return [
        {"content": doc, "metadata": meta}
        for doc, meta in zip(documents, metadatas)
    ]

get_chunks_by_source_and_page('D:/rag_poc/rag_pdfs/proposals/internal\\11-77-63-1616_B.pdf',0)
```
